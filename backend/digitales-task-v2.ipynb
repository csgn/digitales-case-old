# 3. Train Model

```python
ORIGIN_PATH = '/content/drive/MyDrive/digitales-task-v2'
CUSTOM_MODEL_NAME = 'helmet-detector-model'
IMAGES_TRAIN_PATH = '/content/images/train'
IMAGES_TEST_PATH = '/content/images/test'
IMAGES_VAL_PATH = '/content/images/val'
CLASSES = ['helmet', 'vest', 'head']
YAML_PATH = ORIGIN_PATH + '/hard_head.yaml'
YOLO_MODEL = 'yolov8s.pt'
```

```python
import yaml

dict_file = {'train': IMAGES_TRAIN_PATH ,
            'val': IMAGES_VAL_PATH,
            'nc' : len(CLASSES),
            'names' : CLASSES}

with open(YAML_PATH, 'w+') as file:
    documents = yaml.dump(dict_file,file)  
```

```python
from ultralytics import YOLO

model = YOLO(YOLO_MODEL) 
```

```python
model.train(data=YAML_PATH, epochs=20)
```
